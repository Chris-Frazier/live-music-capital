```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import re
import os

# Output File path for data (CSV) and plots (png)
output_fp = "output/"
if not os.path.exists(output_fp):
    os.mkdir(output_fp)

year = 2017

cities_data_filepath = '../us_cities_all_clean.csv'
cities_data = pd.read_csv(cities_data_filepath)
```

```python
url = f'https://api.census.gov/data/{year}/pep/population?get=POP,GEONAME&for=place:*&key=85ac64b6b5a9c0901b00329d1ef41f0c53ccfc98'
data = requests.get(url).json()
pop_df = pd.DataFrame(data)
pop_df = pop_df.drop(index=0, axis=0)
pop_df.columns=['Population', 'Location', 'State Code', 'Place ID']
pop_df

```

```python
locations = []
[locations.append(location) for location in pop_df['Location']]

cities = []
states = []


for location in locations:
    
    location_string_list = re.split(' town, | city, | city \(balance\), | municipality, | city and borough, | borough, | village, | unified government \(balance\), | consolidated government \(balance\), | unified government, | consolidated government, | metropolitan government, | metropolitan government \(balance\), | urban county, | metro government \(balance\), | \(balance\), |, ', location)  
    
    cities.append(location_string_list[0])
    states.append(location_string_list[len(location_string_list)-1])
          
     
pop_df['City'] = cities
pop_df['State'] = states
pop_df

```

```python
target_cities_pop_df = pd.merge(cities_data, pop_df, how='inner', left_on=['city', 'state'], right_on=['City', 'State'])
target_cities_pop_df = target_cities_pop_df.drop(['Location', 'State Code', 'Place ID', 'City', 'State'], 1)
target_cities_pop_df
```

```python
# Save as a csv
# Note to avoid any issues later, use encoding="utf-8"
target_cities_pop_df.to_csv(f"{output_fp}population_us_cities_all_clean_{year}.csv", encoding="utf-8", index=False)
```
