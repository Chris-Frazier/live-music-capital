```python
import requests
import pandas as pd
import numpy as np
import csv
from config import setlist_key
```

### Pull all countries from the Setlist API

```python
countries_url = 'https://api.setlist.fm/rest/1.0/search/countries'

countries_headers = {'Accept': 'application/json',
                     'x-api-key': setlist_key
                    }

countries_response = requests.get(countries_url, headers=countries_headers).json()

countries_response
```

### Loop through countries_response to extract each country code

```python
countries = [x['code'] for x in countries_response['country']]

countries
```

### Find the total number of cities in each country

```python
#Create empty list
cities_totals = []

#Loop through each country code to find data for each country
for country in countries:
    cities_totals_url = 'https://api.setlist.fm/rest/1.0/search/cities'
    cities_totals_params = {'country' : country}

    headers = {'Accept': 'application/json',
               'x-api-key': setlist_key
              }

    #Extract relevant info from the response
    try:
        cities_totals_response = requests.get(cities_totals_url, params=cities_totals_params, headers=headers).json()

        #Find the total number of cities in each country and print the response to keep track of iteration
        countries_total = cities_totals_response['total']
        print(f'{country}: {countries_total}')

        #Find number of total pages a country has to iterate through by
        #dividing total number of cities by 50 (max number of results per page)
        #and then rounding up the result using np.ceil()
        page_count = np.ceil(cities_totals_response['total'] / 50)

        #np.ceil() returns floats, so change the page count numbers to integers
        page_count_int = page_count.astype(int)

        #Append the info to the empty list
        cities_totals.append({'country' : country,
                              'total' : countries_total,
                              'page count' : page_count_int
                             })
    #For countries with no total, print the error in an exception to keep loop going
    except Exception as e:
        print(e)
```

### Extract city data

```python
#Create empty list
cities =[]

#for loop iterates through all pages in each country code to extract city data
for country in cities_totals:

    #Loop through each page in a country code
    for page in range(country['page count']):

        #Make API call for one page
        cities_url = 'https://api.setlist.fm/rest/1.0/search/cities'

        cities_params = {'country' : country['country'],
                         'p' : page+1}

        headers = {'Accept': 'application/json',
                   'x-api-key': setlist_key
                  }

        cities_response = requests.get(cities_url, params=cities_params, headers=headers).json()

        #x is set to help keep count when a result is pulled
        x=1

        #Loop through each city (up to 50) for the current page of results 
        for city in cities_response['cities']:

            #Pull city name, state name, state code, country code, and country name
            #and append to the cities list 
            #(state name and code commented out here so even cities with a name of
            #'' or with no 'state' parameter would be extracted)
            try: 
                city_name = city['name']
#                state = city['state']
#                state_code = city['stateCode']
                country_code = city['country']['code']
                country_name = city['country']['name']
                cities.append({'city' : city_name,
#                               'state' : state,
#                               'state code' : state_code,
                               'country code' : country_code,
                               'country' : country_name
                              })
                print(f"{country['country']}, page {page+1}: {city_name}, {x} of {len(cities_response['cities'])} complete.")
                x+=1
            except Exception as e:
                print(f"{country['country']}, page {page+1}: {city_name}, {type(e)}, {e}, at {x} of {len(cities_response['cities'])}.")
                x+=1
```

### Write data to .csv file

```python
with open('cities_complete_no_states.csv', 'w', newline='', encoding='utf-8') as cities_complete_file:
    cities_file = csv.writer(cities_complete_file)
    cities_file.writerow(cities[0].keys())
    for x in range(len(cities)):
        cities_file.writerow(cities[x].values())
```
