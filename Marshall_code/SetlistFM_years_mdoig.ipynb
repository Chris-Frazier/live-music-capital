```python
import requests
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
from config import setlist_key
```

```python
for year in range(1800, 2020, 1):
    try:
        url = 'https://api.setlist.fm/rest/1.0/search/setlists'

        #params = {'cityName' : 'Austin',
        #          'stateCode' : 'TX',
        #          'countryCode' : 'US'
        #         }

        params = {'year' : year}

        headers = {'Accept': 'application/json',
                   'x-api-key': setlist_key
                  }

        response = requests.get(url, params=params, headers=headers).json()

        total = response['total']

        print(f"{year} - TRUE: There were {total} concerts in {year}.")
    except Exception as e:
        print(f"{year} - FALSE: There were no concerts in {year}.")
```

```python
years_setlists_total = []
for year in range(1899, 2020, 1):
    try:
        url = 'https://api.setlist.fm/rest/1.0/search/setlists'

        params = {'year' : year
                 }

        headers = {'Accept': 'application/json',
                   'x-api-key': setlist_key
                  }

        response = requests.get(url, params=params, headers=headers).json()

        total = response['total']

        years_setlists_total.append({'year': year,
                                     'total' : total
                                    })
        print(f"There were {total} concerts in {year}.")
    except Exception as e:
        print(f"There were no concerts in {year}.")
```

```python
us_top_50_cities = []
with open('../setlists_us_top_50.csv', 'r', newline='', encoding='utf-8') as us_top_50_file:
    us_top_50 = csv.DictReader(us_top_50_file)
    for row in us_top_50:
        us_top_50_cities.append({'cityName' : row['city'],
                                 'stateCode' : row['stateCode'],
                                 'countryCode' : row['countryCode'],
                                })
```

```python
us_top_50_cities
```

```python
last_five_years = [2014, 2015, 2016, 2017, 2018]

us_last_five_years = []
for year in last_five_years:
    top_50_cities_totals = []
    for city in us_top_50_cities:
        try:
            url = 'https://api.setlist.fm/rest/1.0/search/setlists'

            params = {'year' : year,
                      'cityName' : city['cityName'],
                      'stateCode' : city['stateCode'],
                      'countryCode' : city['countryCode']
                     }

            headers = {'Accept': 'application/json',
                       'x-api-key': setlist_key
                      }

            response = requests.get(url, params=params, headers=headers).json()

            total = response['total']

            top_50_cities_totals.append({'city' : city['cityName'],
                                         'stateCode' : city['stateCode'],
                                         'countryCode' : city['countryCode'],
                                         'total' : total
                                        })
        except Exception as e:
            print(f"{type(e)}, {e}")
            
    us_last_five_years.append({year: top_50_cities_totals})
        
    print(f"{year} complete.")
```

```python
us_last_five_years_2014_df = pd.DataFrame(us_last_five_years[0][2014])

us_last_five_years_2015_df = pd.DataFrame(us_last_five_years[1][2015])

us_last_five_years_2016_df = pd.DataFrame(us_last_five_years[2][2016])

us_last_five_years_2017_df = pd.DataFrame(us_last_five_years[3][2017])

us_last_five_years_2018_df = pd.DataFrame(us_last_five_years[4][2018])

us_last_five_years_all_df = pd.merge(us_last_five_years_2014_df, us_last_five_years_2015_df,
                                     how='outer', on=['city', 'stateCode', 'countryCode'], suffixes=('_2014', '_2015'))

us_last_five_years_all_df = pd.merge(us_last_five_years_all_df, us_last_five_years_2016_df,
                                     how='outer', on=['city', 'stateCode', 'countryCode'])

us_last_five_years_all_df = pd.merge(us_last_five_years_all_df, us_last_five_years_2017_df,
                                     how='outer', on=['city', 'stateCode', 'countryCode'], suffixes=('_2016', '_2017'))

us_last_five_years_all_df = pd.merge(us_last_five_years_all_df, us_last_five_years_2018_df,
                                     how='outer', on=['city', 'stateCode', 'countryCode'])

us_last_five_years_all_df = us_last_five_years_all_df.rename(columns={'total' : 'total_2018'})

us_last_five_years_all_df = us_last_five_years_all_df[['city', 'stateCode', 'countryCode', 'total_2014',
                                                       'total_2015', 'total_2016', 'total_2017', 'total_2018']]

us_last_five_years_all_df
```

```python
world_top_50_cities = []
with open('../setlists_world_top_50.csv', 'r', newline='', encoding='utf-8') as world_top_50_file:
    world_top_50 = csv.DictReader(world_top_50_file)
    for row in world_top_50:
        world_top_50_cities.append({'cityName' : row['city'],
                                    'country' : row['country'],
                                    'countryCode' : row['countryCode'],
                                   })
```
