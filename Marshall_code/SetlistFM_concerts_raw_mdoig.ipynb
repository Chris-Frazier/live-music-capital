```python
import requests
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
from config import setlist_key
```

```python
world_cities_complete_df = pd.read_csv('../world_cities_complete.csv')

world_cities_complete_df = world_cities_complete_df.drop_duplicates()

world_cities_complete_df.head()
```

```python
world_cities_no_states_df = pd.read_csv('../world_cities_complete_no_states.csv')

world_cities_no_states_df = world_cities_no_states_df.drop_duplicates()

world_cities_no_states_df.head()
```

```python
world_cities_all_df = pd.merge(world_cities_complete_df, world_cities_no_states_df, how='outer', on=['city', 'country', 'country code'])

world_cities_all_df = world_cities_all_df.drop_duplicates(subset=['city', 'country code', 'country'])

world_cities_all_df = world_cities_all_df.sort_values(by='country code').reset_index(drop=True)

world_cities_all_df = world_cities_all_df.dropna(subset=['city']).reset_index(drop=True)

world_cities_all_df = world_cities_all_df.drop(columns=['state', 'state code'])

world_cities_all_df = world_cities_all_df.fillna('NA')

world_cities_all_df = world_cities_all_df.sort_values(by='country code').reset_index(drop=True)

world_cities_all_df
```

```python
us_cities_complete_df = pd.read_csv('../us_cities_complete.csv')

us_cities_complete_df = us_cities_complete_df.drop_duplicates()

us_cities_complete_df = us_cities_complete_df.sort_values(by='state').reset_index(drop=True)

us_cities_complete_df
```

```python
setlists_us_cities = []
with open('../us_cities_all_clean.csv', 'r', newline='', encoding='utf-8') as us_cities_all_file:
    us_cities_file = csv.DictReader(us_cities_all_file)
    for row in us_cities_file:
        print(f"{row['city']}, {row['state code']}, {row['country code']}")
        setlists_us_cities.append({'cityName' : row['city'],
                                   'stateCode' : row['state code'],
                                   'countryCode' : row['country code'],
                                  })
```

```python
setlists_us_cities
```

```python
setlists_us_cities[0]['cityName']
```

```python
setlists_us_total = []
for city in setlists_us_cities:    
    try:
        url = 'https://api.setlist.fm/rest/1.0/search/setlists'

        params = {'cityName' : city['cityName'],
                  'stateCode' : city['stateCode'],
                  'countryCode' : city['countryCode']
                 }

        headers = {'Accept': 'application/json',
                             'x-api-key': setlist_key
                            }

        response = requests.get(url, params=params, headers=headers).json()

        total = response['total']

        setlists_us_total.append({'city': city['cityName'],
                                  'stateCode' : city['stateCode'],
                                  'countryCode' : city['countryCode'],
                                  'total' : total
                                 })
        print(f"{city['cityName']}, {city['stateCode']} total found.")
    except Exception as e:
        print(f"{type(e)}, {e} for {city['cityName']}, {city['stateCode']}.")
```

```python
setlists_us_total_df = pd.DataFrame(setlists_us_total)

setlists_us_total_df = setlists_us_total_df[['city', 'stateCode', 'countryCode', 'total']]

setlists_us_total_df
```

```python
setlists_world_cities = []
with open('../world_cities_all_clean.csv', 'r', newline='', encoding='utf-8') as world_cities_all_file:
    world_cities_file = csv.DictReader(world_cities_all_file)
    for row in world_cities_file:
        print(f"{row['city']}, {row['country code']}, {row['country']}")
        setlists_world_cities.append({'cityName' : row['city'],
                                      'countryCode' : row['country code'],
                                      'country' : row['country'],
                                     })
```

```python
setlists_world_total = []
for city in setlists_world_cities:    
    try:
        url = 'https://api.setlist.fm/rest/1.0/search/setlists'

        params = {'cityName' : city['cityName'],
                  'countryCode' : city['countryCode']
                 }

        headers = {'Accept': 'application/json',
                   'x-api-key': setlist_key
                  }

        response = requests.get(url, params=params, headers=headers).json()

        total = response['total']

        setlists_world_total.append({'city': city['cityName'],
                                     'countryCode' : city['countryCode'],
                                     'country' : city['country'],
                                     'total' : total
                                    })
        print(f"{city['cityName']}, {city['countryCode']}, {city['country']} total found.")
    except Exception as e:
        print(f"{type(e)}, {e} for {city['cityName']}, {city['countryCode']}, {city['country']}.")
```
